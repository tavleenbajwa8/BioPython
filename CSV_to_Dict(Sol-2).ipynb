```python
import pprint 

#Defining the output format
output_format = dict({"name":"BRCA2","transcripts":[]})   


#Opening the csv file to be converted into json
with open(r"C:\\Users\\Lakshya\\Downloads\\transcripts_ENSG00000139618.csv") as f_b:
    content = f_b.read()

#Removing "\n" using split function
f = content.split("\n")

#Making a list of all rows in a csv (x is a main list, containing other list elements)
x = []
for i in f:
    x.append(i.split(','))
x.pop(-1)
#print(x)

#Assigning keys using first row elements 
headers = x[0]

#Taking lists to be assigned as values from main list x starting from index 1: (last)
vals = x[1:]

#Zipping the lists from vals with main header lists containing common key names
for k in vals:
    d = dict(zip(headers,k))
    output_format["transcripts"].append(d)    #Appending the zipped dictionary to the transcripts list in output format

pprint.pprint(output_format)           #Printing output_format using pretty print
```
