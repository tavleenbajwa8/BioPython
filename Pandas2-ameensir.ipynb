```python
#Partial listing of Series string methods
```

```python
1. Cat: Concatenate strings element wise with optional delimiter
2. contains: Returns boolean array if each string contains pattern/regex
3. count: Count occurences of the pattern

    
4. extract: Use a regex with groups to extract one or more strings from a series of strings; result will be a dataframe with 1 column per group
5. endswith : Equivalent to x.endswith(pattern) for each element
6. startswith : Equivalent to x.startswith(pattern) for each element

7. findall: Compute list of all occurences of a pattern for each string
8. match: Use re.match with the passed regex on each element (returning TRUE/FALSE)

9. get: Index into each element(retrieve ith element)
10. isalnum, isalpha, isdecimal, isdigit, islower, isnumeric, isupper
Equivalent to built-in str.isalnum, str.isalpha.......

```

```python
1. join: Join strings in each element of the series with passed seperator
2. len: Compute length of each string
3. lower, upper: Convert cases (x.lower() or x.upper()) 
    
4. pad: Add whitespace to left, right or both sides of the strings
5. center: Equivalent to pad(side="both")
6. repeat: Duplicate values (s.str.repeat(3) is equivalent to x*3 for each string)
    
7. replace: Replace occurences of pattern/regex with some other string
9. slice: Slice each string in the series
10. split: Split strings on delimiter or regex
11. strip: Trim whitespaces on both sides(inclusing newlines)
12. rstrip: Trim whitespace on right
13. lstrip: Trim whitespace on left
    

    
```

Different join types with how argument
Option Behavior
how="inner" Use only the key combinations observed in both tables
how="left" Use all key combinations found in the left table
how="right" Use all key combinations found in the right table
how="outer" Use all key combinations observed in both tables together
